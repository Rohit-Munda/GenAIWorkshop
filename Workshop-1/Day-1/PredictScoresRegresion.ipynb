# Predict exam scores based on hours of study using Linear Regression

```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Step 2: Sample data (Hours studied vs Exam score)
# X = Hours studied, y = Exam score
X = np.array([[1], [2], [3], [4], [5], [6]])  # Independent variable
y = np.array([35, 45, 50, 60, 65, 75])       # Dependent variable
```

```python
# Step 3: Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 4: Make a prediction
hours_studied = [[7]]
predicted_score = model.predict(hours_studied)
print(f"Predicted exam score for studying {hours_studied[0][0]} hours: {predicted_score[0]:.2f}")
```

```python
# Step 5: Visualize the data and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(hours_studied, predicted_score, color='green', label='Prediction')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
