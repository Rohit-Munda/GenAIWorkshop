```python
#Import necessary packages
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# A small, fast, all-purpose model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "I love playing football on weekends.",
    "Python is a great programming language.",
    "The capital of France is Paris.",
    "Artificial Intelligence is the future.",
    "My dog loves to chase squirrels.",
    "Let's order some pizza tonight.",
    "He plays the guitar very well.",
    "Machine learning helps computers learn from data."
]
```

```python
# Convert documents to vector embeddings
document_embeddings = model.encode(documents)
```

```python
document_embeddings
```

```python
# Dimension of embeddings
embedding_dim = document_embeddings.shape[1]
```

```python
embedding_dim
```

```python
# Create index for cosine similarity search
index = faiss.IndexFlatL2(embedding_dim)
```

```python
# Add the document embeddings to the index
index.add(np.array(document_embeddings))
```

```python
def search_similar(query, top_k=3):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), top_k)

    print(f"\n🔎 Query: {query}\n")
    for i, idx in enumerate(indices[0]):
        print(f"Top {i+1} match: {documents[idx]} (Distance: {distances[0][i]:.4f})")
```

```python
search_similar("What is the capital of France?")
```

```python
search_similar("I want to eat something tasty.")
```

```python
search_similar("Tell me about programming.")
```
